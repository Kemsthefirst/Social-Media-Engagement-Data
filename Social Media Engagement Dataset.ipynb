## Patterns , Trends, And Anomalies

```python
weekday_map = {1: 'Monday', 2: 'Tuesday', 3: 'Wednesday', 4: 'Thursday', 5: 'Friday', 6: 'Saturday', 7: 'Sunday'}

Media['Post Weekday'] = Media['Post Weekday'].map(weekday_map)

weekday_engagement = Media.groupby('Post Weekday')[['like', 'comment', 'share']].sum()

weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

weekday_engagement = weekday_engagement.reindex(weekday_order)

plt.figure(figsize=(10,6))
weekday_engagement.plot(kind='bar')
plt.title('Total Engagement by Weekday')
plt.ylabel('Engagement Count')
plt.xlabel('Weekday')
plt.xticks(rotation=45)
plt.show()
print("Total Engagement by Weekday:")
print(weekday_engagement)
```

```python
Media['Post Month'] = Media['Post Month'].astype(str).apply(lambda x: x.zfill(2))  

monthly_engagement = Media.groupby(Media['Post Month']).agg({
    'like': 'sum',
    'comment': 'sum',
    'share': 'sum'
}).reset_index()

monthly_engagement['Post Month'] = pd.to_datetime(monthly_engagement['Post Month'], format='%m').dt.month
monthly_engagement = monthly_engagement.sort_values('Post Month')

plt.figure(figsize=(10,6))
sns.lineplot(x='Post Month', y='like', data=monthly_engagement, label='Likes', marker='o')
sns.lineplot(x='Post Month', y='comment', data=monthly_engagement, label='Comments', marker='o')
sns.lineplot(x='Post Month', y='share', data=monthly_engagement, label='Shares', marker='o')
plt.title('Engagement Trends Over Time')
plt.xlabel('Month')
plt.ylabel('Engagement Count')
plt.xticks(ticks=range(1, 13), labels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.legend()
plt.show()
print("Total Engagement by Month:")
print(monthly_engagement)

```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=Media[['like', 'comment', 'share']])
plt.title('Outliers in Engagement (Likes, Comments, Shares)')
plt.show()
engagement_summary = Media[['like', 'comment', 'share']].describe()
print("Statistical Summary for Engagement (Likes, Comments, Shares):")
print(engagement_summary)
```
